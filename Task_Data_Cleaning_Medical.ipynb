```python
import pandas as pd
```

```python
#load Dataset
df = pd.read_csv('/content/Medical_Appointment.csv')
#data Overview
df.info()
df.head()
```

```python
#Identify and handle missing values
print("missing values")
print(df.isnull().sum())
#if there any null values then using following command we can Drop rows with missing values
df = df.dropna()
```

```python
#Remove duplicate rows
df = df.drop_duplicates()
```

```python
#Clean and Standardize Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-','_')
```

```python
# Standardize Text Values
df['gender'] = df['gender'].str.strip().str.upper()
df['no_show'] = df['no_show'].str.strip().str.upper()
#converting "No" and "Yes" to Boolean
df['no_show'] = df['no_show'].replace({'NO':0, 'YES': 1})
```

```python
#Convert date formats to a consistent type (e.g., dd-mm-yyyy)
df['appointmentday'] = pd.to_datetime(df['appointmentday'], format='%d-%m-%Y')
df['scheduledday'] = pd.to_datetime(df['scheduledday'], format='%d-%m-%Y')
```

```python
df.head()
```

```python
#Fix Data Types
df['age'] = df['age'].astype(int)
df['patientid'] = df['patientid'].astype(str)
df['appointmentid'] = df['appointmentid'].astype(str)
```

```python
df.info()
```

```python
#Remove negative or unrealistic ages
df = df[df['age'].between(0,115)]
```

```python
#cleaned Data
df.to_csv("Cleaned_Medical_Appointment.csv",index= False)
print("Cleaned_Medical_Appointment.csv downloaded Successfully...")
```
